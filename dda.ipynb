```python
import matplotlib.pyplot as plt
def dda_line(x1, y1, x2, y2):
    points = []

    dx = x2 - x1
    dy = y2 - y1

    steps = int(max(abs(dx), abs(dy)))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(steps + 1):
        points.append((round(x), round(y)))
        x += x_inc
        y += y_inc

    return points
```

```python
# Input from user
x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))
```

```python
dda_points = dda_line(x1, y1, x2, y2)

dda_x, dda_y = zip(*dda_points)

# Plotting
plt.figure(figsize=(8, 8))
plt.plot([x1, x2], [y1, y2], 'r-', label="Actual Line")        # Red actual line
plt.scatter(dda_x, dda_y, c='blue', s=25, label="DDA Points")  # Blue DDA dots

# Make it look like a grid
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, which='both', color='lightgray', linestyle='--', linewidth=0.5)
plt.xticks(range(min(dda_x)-5, max(dda_x)+5, 5))
plt.yticks(range(min(dda_y)-5, max(dda_y)+5, 5))
plt.legend()
plt.title("DDA's Line Drawing Algorithm using Matplotlib")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.figure(figsize=(10, 16))

# Ideal (actual) line as dashed gray
plt.plot([x1, x2], [y1, y2], linestyle='--', color='gray', label='Ideal Line')

# DDA points as red squares
plt.plot(dda_x, dda_y, 'rs', label='DDA Line', markersize=5)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, which='both', color='lightgray', linestyle='--', linewidth=0.5)

plt.xticks(range(min(dda_x) - 5, max(dda_x) + 10, 10))
plt.yticks(range(min(dda_y) - 5, max(dda_y) + 10, 10))

plt.legend()
plt.title("DDA Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
def bresenham_line(x0, y0, x1, y1):
    # Step 1: Save left endpoint as (x0, y0)
    if x0 > x1:  # Ensure drawing left to right
        x0, y0, x1, y1 = x1, y1, x0, y0

    # Step 2: Plot the first point
    points = [(x0, y0)]

    # Step 3: Calculate the constants
    dx = x1 - x0
    dy = y1 - y0
    two_dy = 2 * abs(dy)
    two_dy_minus_dx = 2 * abs(dy) - dx

    # Step 4: Initial decision parameter
    p = two_dy - dx

    # Step 5: Set initial point
    x, y = x0, y0
    y_step = 1 if dy > 0 else -1  # handle upward/downward slope

    # Step 6: Loop through x and determine each pixel
    for _ in range(dx):
        x += 1
        if p < 0:
            p += two_dy
        else:
            y += y_step
            p += two_dy_minus_dx
        points.append((x, y))

    # Step 7: Done
    return points
```

```python
x1, y1 = int(input("Enter start x: ")), int(input("Enter start y: "))
x2, y2 = int(input("Enter end x: ")), int(input("Enter end y: "))

# Get Bresenham line points
bresenham_points = bresenham_line(x1, y1, x2, y2)
x_coords, y_coords = zip(*bresenham_points)
```

```python
# Plot the result using matplotlib
plt.figure(figsize=(6, 7))
plt.plot([x1, x2], [y1, y2], label="Ideal Line", color="gray", linestyle="--")
plt.scatter(x_coords, y_coords, c="black", label="Bresenham Pixels", s=20)

# Make it look like a grid
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, which='both', color='lightgray', linestyle='--', linewidth=0.5)
plt.xticks(range(min(x_coords)-5, max(x_coords)+5, 5))
plt.yticks(range(min(y_coords)-5, max(y_coords)+5, 5))
plt.legend()
plt.title("Bresenham's Line Drawing Algorithm using Matplotlib")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Plot using normal plt.plot() with square markers
plt.figure(figsize=(10, 16))
plt.plot([x1, x2], [y1, y2], linestyle='--', color='gray', label='Ideal Line')  # ideal line
plt.plot(x_coords, y_coords, 'rs', label="Bresenham Line", markersize=5)  # square markers

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, which='both', color='lightgray', linestyle='--', linewidth=0.5)
plt.xticks(range(min(x_coords)-5, max(x_coords)+10, 10))
plt.yticks(range(min(y_coords)-5, max(y_coords)+10, 10))
plt.legend()
plt.show()
```

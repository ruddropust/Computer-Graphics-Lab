```python
import matplotlib.pyplot as plt
def bresenham_line(start,end):
    dx=abs(end[0]-start[0])
    dy=abs(end[1]-start[1])
    p0=2*dy-dx
    activated_pixels=[]
    x=start[0]
    y=start[1]
    activated_pixels.append((x,y))
    while x!=end[0] or y!=end[1]:
        if p0<1:
            x=x+1
            p0=p0+2*dy
        else:
            x=x+1
            y=y+1
            p0=p0+2*dy-2*dx
        activated_pixels.append((x,y))
    return activated_pixels
def plot_point(activated_pixel):
    x,y=zip(*activated_pixel)
    plt.plot(x,y)
    plt.show()
    
x1=int(input("Enter x1:"))
y1=int(input("Enter y1:"))
x2=int(input("Enter x2:"))
y2=int(input("Enter y2:"))
activated_pixel=bresenham_line((x1,y1),(x2,y2))
print(activated_pixel)
plot_point(activated_pixel)
        
```

```python
import matplotlib.pyplot as plt
import math
def rotation(x,y,angel_deg):
    angel_rad=math.radians(angel_deg)
    x1=x*math.cos(angel_rad)-y*math.sin(angel_rad)
    y1=x*math.sin(angel_rad)+y*math.cos(angel_rad)
    return [x1,y1]

def translation(x,y,tx,ty):
    x=x+tx
    y=y+ty
    return [x,y]
def scaling(x,y,sx,sy):
    x=x*sx
    y=y*sy
    return [x,y]


def shear(x,y,shx,shy):
    x1=x+shx*y
    y1=y+shy*x
    return [x1,y1]
    
    
def reflection(x,y,mode):
    if mode=='x':
        y=-y
    elif mode=='y':
        x=-x
    elif mode=='o':
        x=-x
        y=-y
    else:
        print("Invalid reflection mode")
    return [x,y]

def pivot_rotation(x,y,xp,yp,angle_deg):
    point1=translation(x,y,-xp,-yp)
    point2=rotation(point1[0],point1[1],angle_deg)
    output=translation(point2[0],point2[1],xp,yp)    
    return output

def plot_shapes(original, transformed, label1, label2, title):
    x_orig, y_orig = zip(*(original + [original[0]]))
    x_trans, y_trans = zip(*(transformed + [transformed[0]]))
    plt.figure(figsize=(7,7))
    plt.plot(x_orig, y_orig, marker='o', label=label1)
    plt.plot(x_trans, y_trans, marker='o', label=label2)
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable="box")
    plt.legend()
    plt.title(title)
    plt.show()
```

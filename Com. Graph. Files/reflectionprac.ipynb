```python
import matplotlib.pyplot as plt
import math
def reflection(x,y,mode):
   if mode=='x':
       y=-y
   elif mode =='y':
       x=-x
   elif mode =='o':
       x=-x
       y=-y
   return [x,y]
def plot(points):
    for i in range(len(points)-1):
        x1=points[i][0]
        x2=points[i+1][0]
        y1=points[i][1]
        y2=points[i+1][1]
        plt.plot([x1,x2],[y1,y2])
        
n=int(input("Enter the shape:"))
point=[]
for i in range(n):
    x=int(input (f"enter {i+1}-x:"))
    y=int(input (f"enter {i+1}-y:"))
    point.append([x,y])
mode=input("Enter reflection mode(x/y/o+):")
reflection_point=[]
for  i in range(len(point)):
    x=point[i][0]
    y=point[i][1]
    new_point=reflection(x,y,mode)
    reflection_point.append(new_point)
plot(point)
plot(reflection_point)
plt.show()

        
        
        
```

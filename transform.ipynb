```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Function to draw shapes
def draw_shape(original, transformed, title="Transformation"):
    plt.figure()
    plt.plot(original[:, 0], original[:, 1], 'bo-', label='Original')
    plt.plot(transformed[:, 0], transformed[:, 1], 'ro-', label='Transformed')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Transformation functions
def translate(shape, tx, ty):
    return shape + np.array([tx, ty])

def scale(shape, sx, sy):
    return shape @ np.array([[sx, 0], [0, sy]])

def rotate(shape, angle_deg):
    angle_rad = math.radians(angle_deg)
    rot_mat = np.array([[math.cos(angle_rad), -math.sin(angle_rad)],
                        [math.sin(angle_rad),  math.cos(angle_rad)]])
    return shape @ rot_mat

def reflect(shape, axis='x'):
    if axis == 'x':
        ref_mat = np.array([[1, 0], [0, -1]])
    elif axis == 'y':
        ref_mat = np.array([[-1, 0], [0, 1]])
    elif axis == 'origin':
        ref_mat = np.array([[-1, 0], [0, -1]])
    elif axis == 'y=x':
        ref_mat = np.array([[0, 1], [1, 0]])
    else:
        print("Invalid axis.")
        return shape
    return shape @ ref_mat

def shear(shape, shx=0, shy=0):
    shear_mat = np.array([[1, shx], [shy, 1]])
    return shape @ shear_mat

# Function to take shape input from user
def get_shape_from_user():
    n = int(input("Enter number of vertices: "))
    coords = []
    for i in range(n):
        x = float(input(f"Enter x{i+1}: "))
        y = float(input(f"Enter y{i+1}: "))
        coords.append([x, y])
    coords.append(coords[0])  # Closing the shape
    return np.array(coords)

# Menu
def menu():
    shape = get_shape_from_user()

    while True:
        print("\n--- Menu ---")
        print("1. Translation")
        print("2. Scaling")
        print("3. Rotation")
        print("4. Reflection")
        print("5. Shearing")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            tx = float(input("Enter tx: "))
            ty = float(input("Enter ty: "))
            transformed = translate(shape, tx, ty)
            draw_shape(shape, transformed, "Translation")

        elif choice == '2':
            sx = float(input("Enter sx: "))
            sy = float(input("Enter sy: "))
            transformed = scale(shape, sx, sy)
            draw_shape(shape, transformed, "Scaling")

        elif choice == '3':
            angle = float(input("Enter rotation angle (degrees): "))
            transformed = rotate(shape, angle)
            draw_shape(shape, transformed, "Rotation")

        elif choice == '4':
            axis = input("Reflect over (x / y / origin / y=x): ").lower()
            transformed = reflect(shape, axis)
            draw_shape(shape, transformed, f"Reflection over {axis}")

        elif choice == '5':
            shx = float(input("Enter shear in X (shx): "))
            shy = float(input("Enter shear in Y (shy): "))
            transformed = shear(shape, shx, shy)
            draw_shape(shape, transformed, "Shearing")

        elif choice == '6':
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Please try again.")

# Run the menu
menu()
```
